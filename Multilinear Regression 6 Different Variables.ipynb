```python
import pandas as pd
from sklearn import linear_model
import tkinter as tk 
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

kc_data = pd.read_csv("kc_house_data.csv")

df = pd.DataFrame(kc_data,columns=['price','condition', 'bathrooms', 'view', 'sqft_living', 'grade', 'waterfront']) 


X = df[['condition','bathrooms', 'view', 'sqft_living', 'grade', 'waterfront']].astype(float) # here we have 2 input variables for multiple regression. If you just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] for example.Alternatively, you may add additional variables within the brackets
Y = df['price'].astype(float) # output variable (what we are trying to predict)

df.head()
```

```python
kc_data_del_outlier = kc_data[kc_data.bedrooms != 33]
df = pd.DataFrame(kc_data_del_outlier,columns=['price','bedrooms', 'bathrooms', 'yr_built', 'sqft_living', 'grade', 'waterfront'])
```

```python


# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

# tkinter GUI
root= tk.Tk()

canvas1 = tk.Canvas(root, width = 500, height = 300)
canvas1.pack()

# with sklearn
Intercept_result = ('Intercept: ', regr.intercept_)
label_Intercept = tk.Label(root, text=Intercept_result, justify = 'center')
canvas1.create_window(300, 220, window=label_Intercept)

# with sklearn
Coefficients_result  = ('Coefficients: ', regr.coef_)
label_Coefficients = tk.Label(root, text=Coefficients_result, justify = 'center')
canvas1.create_window(300, 240, window=label_Coefficients)

# Bedrooms label and input box
label1 = tk.Label(root, text='Condition (1-5): ')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)

# Bathrooms label and input box
label2 = tk.Label(root, text=' Type number of bathrooms: ')
canvas1.create_window(120, 120, window=label2)

entry2 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 120, window=entry2)

# Year Build label and input box
label3 = tk.Label(root, text='View(0/1): ')
canvas1.create_window(140, 140, window=label3)

entry3 = tk.Entry (root) # create 3rd entry box
canvas1.create_window(270, 140, window=entry3)


# Squarefoot living Room and input box
label4 = tk.Label(root, text=' Square Foot Living Space: ')
canvas1.create_window(160, 160, window=label4)

entry4 = tk.Entry (root) # create 4th entry box
canvas1.create_window(290, 160, window=entry4)

# Grade and input box
label5 = tk.Label(root, text=' Grade (0-12): ')
canvas1.create_window(180, 180, window=label5)

entry5 = tk.Entry (root) # create 5th entry box
canvas1.create_window(290, 180, window=entry5)

#Waterfront Input Box
label6 = tk.Label(root, text='Waterfront(1/0): ')
canvas1.create_window(200, 200, window=label6)

entry6 = tk.Entry (root) # create 5th entry box
canvas1.create_window(290, 200, window=entry6)


def values(): 
    global condition #our 1st input variable
    condition = float(entry1.get()) 
    
    global new_bathrooms #our 2nd input variable
    new_bathrooms = float(entry2.get()) 
    
    global view #our 3nd input variable year built
    view = float(entry3.get()) 
    
    global sqft_living #our 4th input variable sqft living are
    sqft_living = float(entry4.get())
    
    global grade #our 5th input variable grade
    grade = float(entry5.get())
    
    global waterfront #our 6th input variable waterfront
    waterfront = float(entry6.get())
    
    Prediction_result  = ('Predicted House Price: ', regr.predict([[condition, new_bathrooms, view, sqft_living, grade, waterfront]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(290, 280, window=label_Prediction)
    
button1 = tk.Button (root, text='Predict House Price',command=values, bg='orange') # button to call the 'values' command above 
canvas1.create_window(290, 300, window=button1) ##########
 




root.mainloop()
```

```python

#plot 1st scatter  for bedrooms
figure3 = plt.Figure(figsize=(2,1), dpi=100)
ax3 = figure3.add_subplot(111)
ax3.scatter(df['bedrooms'].astype(float),df['price'].astype(float),  color = 'r')
scatter3 = FigureCanvasTkAgg(figure3, root) 
scatter3.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax3.legend(['price']) 
ax3.set_xlabel('bedrooms')
ax3.set_title('Bedrooms Vs. Price')

#plot 2nd scatter for bathrooms
figure4 = plt.Figure(figsize=(2,1), dpi=100)
ax4 = figure4.add_subplot(111)
ax4.scatter(df['bathrooms'].astype(float),df['price'].astype(float), color = 'g')
scatter4 = FigureCanvasTkAgg(figure4, root) 
scatter4.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax4.legend(['price']) 
ax4.set_xlabel('Bathrooms')
ax4.set_title('Bathrooms Vs. Price')

#plot 3rd scatter for year built
figure5 = plt.Figure(figsize=(2,1), dpi=100)
ax5 = figure5.add_subplot(111)
ax5.scatter(df['yr_built'].astype(float),df['price'].astype(float), color = 'g')
scatter5 = FigureCanvasTkAgg(figure5, root) 
scatter5.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax5.legend(['price']) 
ax5.set_xlabel('Year_built')
ax5.set_title('Year Built Vs. Price')

#plot 4th  scatter for sqft_living area
figure6 = plt.Figure(figsize=(2,1), dpi=100)
ax6 = figure6.add_subplot(111)
ax6.scatter(df['sqft_living'].astype(float),df['price'].astype(float), color = 'g')
scatter6 = FigureCanvasTkAgg(figure6, root) 
scatter6.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax6.legend(['price']) 
ax6.set_xlabel('sqft_living')
ax6.set_title('Square Foot Living Area Vs. Price')
```
